```python
import io
import soundfile as sf
from datasets import Audio

# Disable torchcodec-based decoding
dataset_en = dataset_en.cast_column("reference_audio", Audio(decode=False))

# Access audio data manually
example = dataset_en[0]

# Decode audio bytes using soundfile
audio_bytes = example['reference_audio']['bytes']
array, sampling_rate = sf.read(io.BytesIO(audio_bytes))

print(f"Text: {example['text']}")
print(f"Audio sampling rate: {sampling_rate}")
print(f"Audio array shape: {array.shape}")
```

```python
# Play audio (if using jupyter/colab)
import IPython.display as ipd
ipd.Audio(example['reference_audio']['array'], rate=example['reference_audio']['sampling_rate'])
```
